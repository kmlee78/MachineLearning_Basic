```python
import FinanceDataReader as fdr

sp500 = fdr.StockListing('S&P500')
sp500.head(10)
```

```python
# 그래프 크기 및 모양 설정
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (14,8)
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["axes.grid"] = True
plt.rcParams['axes.axisbelow'] = True
```

```python
"""디렉토리 생성 후 csv파일로 저장하기"""

import os
folder = 'sp500_price'
# 'sp500_price' 라는 디렉토리(폴더)가 존재하지 않는다면 이 이름으로 하나 만들어줌
if not os.path.isdir(folder):
    os.mkdir(folder)
```

```python
import csv
import pandas as pd

# sp500 데이터가 500개가 넘으니까 일단 10개만 뽑음 
sp500_sample = sp500.iloc[:10]
for ix, row in sp500_sample.iterrows():   # sp500_sample의 모든 row에 대해서
    sym, name = row['Symbol'], row['Name']
    csv_fd = os.path.join(folder, f'{sym}.csv')   # 폴더 속 csv파일 경로에 연결
    if os.path.exists(csv_fd):    # 이미 존재하고 있다면 스킵하고 다음 row로
        continue
    
    try:
        df = fdr.DataReader(sym, '2019')   # 각 기업 symbol에 대한 주가 데이터 가져오기
        df['Symbol'] = sym
        df.to_csv(csv_fd)    # to_csv('경로') 메소드로 csv형태로 저장
    except ValueError as e:
      print(e)
    except IndexError as e:
      print(e) 
```

```python
# 제대로 만들어졌는지 csv파일 하나 읽어보자
sym = 'AAP'
csv_fd = os.path.join(folder, f'{sym}.csv')
aap = pd.read_csv(csv_fd, parse_dates=True, index_col='Date')
aap.head()
```

```python
# 두 그래프 비교
aapl = fdr.DataReader('AAPL', '2010-01-01', '2019-12-31')
us500 = fdr.DataReader('US500', '2010-01-01', '2019-12-31')

df_plot = pd.DataFrame()
df_plot['US500'] = us500['Close']
df_plot['AAPL'] = aapl['Close']
# 변화율
df_plot = df_plot/df_plot.iloc[0] - 1
df_plot.plot()
```

```python
"""종목별 가격 데이터프레임 & 수익률"""

df_price = pd.DataFrame()
for ix, row in sp500_sample.iterrows():
    try:
        sym, name = row['Symbol'], row['Name']
        csv_fd = os.path.join(folder, f'{sym}.csv')
        df = pd.read_csv(csv_fd, parse_dates=True, index_col='Date')
        df_price[sym] = df['Close']
    except FileNotFoundError as e:
        print(e)
    except ValueError as e:
        print(e)

print(df_price.columns)
```

```python
returns = (df_price.loc['2019'].pct_change() + 1).product() - 1
returns = returns.sort_values(ascending=False)

best3 = returns.head(3)
worst3 = returns.tail(3)
print(best3)
print(worst3)
```

```python
# 가장 높은 수익률의 기업들
df_plot = pd.DataFrame()
df_plot['S&P500'] = us500.loc['2019', 'Close']
df_plot[best3.keys()] = df_price.loc['2019', best3.keys()]

df_plot = df_plot/df_plot.iloc[0] - 1
df_plot.plot()
```

```python
# 가장 낮은 수익률의 기업들
df_plot = pd.DataFrame()
df_plot['S&P500'] = us500.loc['2019', 'Close']
df_plot[worst3.keys()] = df_price.loc['2019', worst3.keys()]

df_plot = df_plot/df_plot.iloc[0] - 1
df_plot.plot()
```

```python
"""종목간 상관관계"""

df = pd.DataFrame()
df['S&P500'] = us500.loc['2019', 'Close']
df[returns.keys()] = df_price.loc['2019', returns.keys()]
df.head(5)
```

```python
changes = df.pct_change()
corr = changes.corr()
corr
```

```python
# 상관계수 순위

idx, val = [], []
for ix, i in enumerate(corr.columns.values):
    for j in corr.columns.values[ix+1:]:
        idx.append((i, j))
        val.append(corr[i][j])

ser = pd.Series(val, index=idx)
ser = ser.sort_values(ascending=False)
ser.head(10)
```

```python
# 특정 지수와의 상관관계 순위
df_mmm = corr['MMM'].sort_values(ascending=False)[1:]
df_mmm
```
