```python
"""바이앤 홀드 전략: 주식 장기 보유 투자 전략"""

import pandas as pd
import numpy as np

df = pd.read_csv('../data/AMZN.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
# 그래프 크기 미리 설정
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (14,8)

# 결측치 확인
df[df.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
df_price = df.loc[:,['Adj Close']].copy()
df_price.plot()
```

```python
# 일별 수익
df_price['daily_rtn'] = df_price['Adj Close'].pct_change()

# 전략 수익(누적 수익) cumprod()를 통해 일별 수익률을 누적곱 해준다
df_price['st_rtn'] = (df_price['daily_rtn'] + 1).cumprod()
df_price.head()
```

```python
df_price['st_rtn'].plot()
```

```python
# 기준을 다르게 수익률 계산하기

base_date = '2020-09-02'
df_temp = df_price.loc[base_date:,['st_rtn']]/df_price.loc[base_date,['st_rtn']]
last_date = df_temp.index[-1]
print('누적 수익: ', df_temp.loc[last_date, 'st_rtn'])
df_temp.plot(figsize=(16,5))
```

```python
"""성과 분석 지표들"""

# CAGR: 연평균 복리 수익률 (252영업일 = 1년)
CAGR = df_price.loc[last_date,'st_rtn'] ** (252./len(df_price.index)) - 1

# MDD: 최대 낙폭, 투자기간 고점부터 저점까지 떨어진 낙폭 중 최댓값 cummax()는 누적 최댓값 cummin()은 최소
highest = df_price['Adj Close'].cummax()
draw_down = df_price['Adj Close']/highest - 1.0
lowest = draw_down.cummin()
MDD = lowest.min()

# VOL(변동성): 수익률 표준편차 * 루트1년
VOL = np.std(df_price['daily_rtn']) * np.sqrt(252.)

# 샤프 지수: 위험 대비 수익성 지표. 실현 수익률 산술 평균 / 실현 수익률의 변동성
Sharpe = np.mean(df_price['daily_rtn']) / (np.std(df_price['daily_rtn']) * np.sqrt(252.))

print('CAGR: ', round(CAGR*100, 2), '%')
print('MDD: ', -round(MDD*100, 2), '%')
print('VOL', round(VOL*100, 2), '%')
print('Sharpe', round(Sharpe, 2))
```
