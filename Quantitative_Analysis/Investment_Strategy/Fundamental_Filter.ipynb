```python
import FinanceDataReader as fdr
import pandas as pd

info = pd.read_csv('data/Corporation_Info.csv', index_col='Corporation')
info.drop(['Unnamed: 0'], axis=1, inplace=True)
info
```

```python
"""필터1"""

filter1 = (info['PER'] < 5) & (info['PBR'] < 1) & (info['ROE'] > 15) & (info['Debt_R'] < 100) & (info['Operating_R'] > 20)
info[filter1]
```

```python
"""필터2: 모든 종목의 종가를 불러와 리스트에 담은 후, 20일 이동평균선을 계산해 필터링"""

import datetime

date = datetime.date(2019, 4, 1)
today = date.today()

stock_list = []
for index in info.index:
    stock_code = str(info.loc[index, 'Symbol'])
    stock_code = stock_code.zfill(6)   # 종목코드가 6자리가 될 수 있게 0을 채워줌
    stock_list.append([index, stock_code])
    

df_list = [fdr.DataReader(code, '2021-01-01', today)['Close'] for name, code in stock_list]
df = pd.concat(df_list, axis=1)
df.columns = [name for name, code in stock_list] 
df
```

```python
filter2 = []
df_new = df.reset_index(drop=True, inplace=False)
valid = True
check_day = 30   # 확인하려는 최근 일수

for corp in df_new.columns:
    df_new[corp + 'tw'] = df_new[corp].rolling(window=20).mean()
    valid = True
    period = len(df_new) - check_day
    for day in range(period, len(df_new)):
        if df_new.loc[day, corp] < df_new.loc[day, corp + 'tw']:
            valid = False
            break
    
    if valid:
        filter2.append(corp)
        
filter2
```
