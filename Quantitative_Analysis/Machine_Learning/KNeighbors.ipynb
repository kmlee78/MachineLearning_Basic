```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# 주가 데이터를 불러올 수 있는 모듈
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
```

```python
df = pdr.get_data_yahoo('SPY', '2012-01-01', '2015-01-01')
df.head()
```

```python
tmp_df = df[['High', 'Low', 'Open', 'Close']].copy()
tmp_df['Open-Close'] = tmp_df['Open'] - tmp_df['Close']
tmp_df['High-Low'] = tmp_df['High'] - tmp_df['Low']

X = tmp_df[['High-Low', 'Open-Close']]
# 다음날의 종가가 다음날의 시가 보다 크면 1, 반대면 -1
y = np.where(tmp_df['Close'].shift(-1) > tmp_df['Open'].shift(-1), 1, -1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False, random_state=3)
```

```python
# 모델 학습. k=1 부터 k=15 까지 학습된 모델을 리스트 안에 넣어준다.
train_acc = []
test_acc = []

for k in range(1, 16):
    clf = KNeighborsClassifier(n_jobs=-1, n_neighbors=k)
    clf.fit(X_train, y_train)
    prediction = clf.predict(X_test)
    train_acc.append(clf.score(X_train, y_train))
    test_acc.append((prediction==y_test).mean())   # y_test랑 예측한 값이 서로 얼마나 맞는지
```

```python
plt.figure(figsize=(12,9))
plt.plot(range(1, 16), train_acc, label='Train set')
plt.plot(range(1, 16), test_acc, label='Test set')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.xticks(np.arange(0, 16, step=1)) 
plt.legend
```

```python
# k = 16일 때 모델을 학습하고 해당 모델로 바이앤홀드 전략 구사했을 때 수익률 그래프 및 샤프지수도 계산
knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(X_train, y_train)

accuracy_train = accuracy_score(y_train, knn.predict(X_train))
accuracy_test = accuracy_score(y_test, knn.predict(X_test))

# cumsum(): 누적 합
tmp_df['Predicted_Signal'] = knn.predict(X)
tmp_df['SPY_ret'] = np.log(tmp_df['Close'] / tmp_df['Open'])
cum_spy_ret = tmp_df['SPY_ret'].cumsum() * 100

tmp_df['Strategy_ret'] = tmp_df['SPY_ret'] * tmp_df['Predicted_Signal'].shift(1)
cum_st_ret = tmp_df['Strategy_ret'].cumsum() * 100

plt.figure(figsize=(12,6))
plt.plot(cum_spy_ret, color='b',label='spy ret')
plt.plot(cum_st_ret , color='r', label='strategy ret')
plt.legend()
plt.show()
```

```python
std = cum_st_ret.std()
sharpe = (cum_st_ret - cum_spy_ret) / std
sharpe = sharpe.mean()
print('Sharpe ratio: ', sharpe)
```
