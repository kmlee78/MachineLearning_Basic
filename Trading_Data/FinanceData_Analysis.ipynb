```python
import pandas as pd

df = pd.read_csv('../data/SPY.csv', index_col='Date', parse_dates=['Date'])
# 행 단위로 데이터 추출
df[0:3]
```

```python
# Date값을 이용해 생 데이터 슬라이싱
df['2020-04-01':'2021-02-01']
```

```python
# loc인덱서 활용
df.loc['2020-05-03':'2020-08-25', ['Open', 'High', 'Low']]
```

```python
# iloc인덱서 활용
df.iloc[200:205,[3,4,5]]
```
