```python
import pandas as pd
df = pd.read_csv('../data/SPY.csv', index_col='Date', parse_dates=['Date'])

# shift()함수: 인덱스에 연결된 데이터를 일정 간격으로 이동시키는 함수 디폴트 파라미터값이 periods=1 이다
df['Close_lag1'] = df['Close'].shift()
df.head()
```

```python
# pct_change()함수: 현재 값과 이전 요소 값의 백분율 변화량
df['pct_change'] = df['Close'].pct_change()
df.head()
```

```python
# diff()함수: 변화량을 구해주는 것(뺄셈)
df['Close_diff'] = df['Close'].diff()
df.head()
```

```python
df = pd.read_csv('../data/SPY.csv', index_col='Date', parse_dates=['Date'])

# rolling()함수: 일정 구간의 데이터들의 평균,최소,최대 등을 계산할 때 쓰인다
# 이동평균선 같은거 구할 때 사용하면 좋다
# window=5는 윈도우 사이즈가 5이기 때문에 처음 4개의 데이터는 nan이 나온다. 그 후에 한 칸씩 내려오면서 함
df['MA'] = df['Close'].rolling(window=5).mean()
df
```

```python
# resample()함수: 시간 간격을 재종하는 기능을 한다
index = pd.date_range('2019-01-01', '2019-10-01', freq='B')
series = pd.Series(range(len(index)), index=index)
series
```

```python
# 각 달의 마지막 날(M)을 기준으로 데이터를 정렬하고, 거기에 sum()을 붙여 월별 합계를 구할 수 있다
series.resample(rule='M').sum()
```

```python
# 월말 일자만 확인
series.resample(rule='M').last()
```

```python
# 월초 일자만 확인
series.resample(rule='M').first()
```
